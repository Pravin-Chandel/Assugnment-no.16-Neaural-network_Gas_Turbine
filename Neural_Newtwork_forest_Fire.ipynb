# 5.Model Building 

```python
X=Forest_data_2.iloc[:,0:10]
y=Forest_data_2.iloc[:,-1].values
```

# 6. Normalizing the data for better results

```python
from sklearn.model_selection import train_test_split

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
X_norm = norm_func(X)
```

# 7. Splitting the data into Training & Testing

```python
X_train,X_test,y_train,y_test = train_test_split(X_norm,y, test_size = 0.2,stratify = y)
```

# 8. Applying Neural network

```python
pip install keras
```

```python
pip install tensorflow
```

```python
import keras.models
import tensorflow
from tensorflow.keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
model.add(Dense(12, input_dim = 10, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8,  kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1,  kernel_initializer = 'uniform', activation = 'linear'))
```

```python
model.compile(loss = 'mse', optimizer = 'adam', metrics = ['accuracy'])
```

# Model Training 

```python
model.fit(X_train,y_train, validation_split = 0.33, epochs = 100, batch_size = 10)
```

```python
# list all data in history
model.history.history.keys()
```

### Let's See the accuracy & Loss of both Trained model & Tested Model

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc = 'upper left')
plt.show()
```

#  Tuning of Hyperparameter : Batch Size and Epochs
 

```python
from sklearn.model_selection import GridSearchCV, KFold
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
from tensorflow.keras.models import load_model
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(12, input_dim = 10, kernel_initializer = 'uniform', activation = 'relu'))
    model.add(Dense(8,  kernel_initializer = 'uniform', activation = 'relu'))
    model.add(Dense(1,  kernel_initializer = 'uniform', activation = 'linear'))
    
    adam = Adam(lr = 0.01)
    model.compile(loss = 'mse', optimizer = adam, metrics = ['accuracy'])
    return model
```

### Create a Model

```python
model_1 = KerasClassifier(build_fn = create_model,verbose = 0)
```

### Defining the Grid Search Parameters

```python
batch_size = [10,20,40]
epochs = [10,50,100]
```

### Making a dictionary of the grid search parameters


```python
Param_grid = dict(batch_size = batch_size,epochs = epochs)
```

### Building and Training the GridSearchCV  Model


```python
grid = GridSearchCV(estimator = model_1, param_grid = Param_grid,cv = KFold(),verbose = 10)
grid_result_Train = grid.fit(X_train,y_train)
```

```python
grid_result_test=grid.fit(X_test,y_test)
```
